```python
df_1 = pd.read_csv('raw_dataset_1_world.csv', sep=';')
df_2 = pd.read_csv('raw_dataset_2_world.csv', sep=';')
df_3 = pd.read_csv('raw_dataset_3_world.csv', sep=';')
df_4 = pd.read_csv('raw_dataset_4_world.csv', sep=';')
```

```python
df = pd.concat([df_1, df_2, df_3, df_4])
df = df[df.gid != "gid"]
df = df[df.gid != "?"]
df = df.drop_duplicates().reset_index(drop=True)
```

```python
cols = df.columns.drop(['gid', 'date'])
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df = df.groupby(['gid', 'date']).sum().reset_index()
df
```

```python
for i in [df_1, df_2, df_3, df_4]:
    del (i)
```

```python
geo_1 = gpd.read_file('resources/gadm_410-levels.gpkg', layer='ADM_1')
geo_2 = gpd.read_file('resources/gadm_410-levels.gpkg', layer='ADM_2')
```

```python
country_l2 = geo_2['GID_0'].unique()
geo_1 = geo_1[~geo_1['GID_0'].isin(country_l2)]

geo_1['GID_2'] = geo_1['GID_1']
geo_1['NAME_2'] = geo_1['NAME_1']

geo_1 = geo_1[['GID_0', 'COUNTRY', 'GID_1', 'NAME_1', 'GID_2', 'NAME_2']]
geo_2 = geo_2[['GID_0', 'COUNTRY', 'GID_1', 'NAME_1', 'GID_2', 'NAME_2']]

geo = pd.concat([geo_1, geo_2])
```

```python
geo_2[geo_2.GID_0 == "USA"]
```

```python
day_month = {'1':'12', '32':'1', '60':'2', '61':'2', '91':'3', '92':'3', '121':'4', '122':'4','152':'5',
             '153':'5', '182':'6', '183':'6','213':'7', '214':'7', '244':'8', '245':'8', '274':'9', '275':'9',
             '305':'10', '306':'10', '335':'11', '336':'11'}

df_data = pd.DataFrame.from_dict(day_month, orient='index').reset_index()
df_data.columns = ['dia', 'mes']
df_data[['mes', 'dia']] = df_data[['mes', 'dia']].apply(np.int64)


df_final = df.merge(geo, how='left', left_on='gid', right_on='GID_2')
df_final[['ano', 'dia']] = df_final['date'].str.split('/', expand=True)
df_final = df_final.drop(['date'], axis=1)
df_final[['ano', 'dia']] = df_final[['ano', 'dia']].apply(np.int64)
df_final = df_final.merge(df_data, on='dia', how='left')

df_final['aux'] = np.select([df_final['mes'] == 12], [1])
df_final['ano'] = df_final['ano'] - df_final['aux']
df_final = df_final.drop(['aux', 'dia', 'gid', 'count_pixel'], axis=1)

df_final = df_final.rename(columns={'GID_0':'country_code', 'COUNTRY':'country_name', 'GID_1':'region_code', 'NAME_1': 'region_name', 'GID_2':'subregion_code', 'NAME_2':'subregion_name', 'ano':'year', 'mes':'month'})
```
